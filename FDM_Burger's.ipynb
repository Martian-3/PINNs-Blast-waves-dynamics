```python
import numpy as np
import matplotlib.pyplot as plt

#nu = 0.00465

# Constants and parameters
nu = 0.00465   # Viscosity (adjustable between 0 and 0.05)
xmax = 1   # Total size of the spatial domain (from -1 to 1)
xmin = -1  # Start of the spatial domain
tmax = 1   # Total time duration
nx = 200   # Number of spatial grid points
dt = 0.001 # Initial temporal resolution

# Discretization
x = np.linspace(xmin, xmax, nx)  # Position vector from -1 to 1
dx = (xmax - xmin) / (nx - 1)    # Spatial resolution
u_max = 1.0                      # Maximum initial velocity amplitude from sin(pi * x)

# Recalculate dt based on the CFL condition (ensuring stability)
dt = min(dt, 0.5 * dx / u_max)   # CFL condition: adjust time step based on max velocity

nt = int(tmax / dt) + 1          # Adjusted to ensure final time step is included
t = np.linspace(0, tmax, nt)     # Time vector

# Initialize the velocity field
u = np.zeros([nx, nt])
u[:, 0] = -np.sin(np.pi * x)     # Initial condition: u(x, t = 0) = -sin(pi * x)
u[0, :] = 0                      # Boundary condition at x = -1
u[-1, :] = 0                     # Boundary condition at x = 1

# Time stepping loop (Explicit finite difference method)
for n in range(nt-1):
    # Apply central difference for space and forward difference for time
    u[1:-1, n+1] = (u[1:-1, n] +
                    dt * (nu * (u[2:, n] - 2 * u[1:-1, n] + u[:-2, n]) / dx**2 -
                          u[1:-1, n] * (u[2:, n] - u[:-2, n]) / (2 * dx)))

    # Clip values to prevent overflow and ensure stability
    u[1:-1, n+1] = np.clip(u[1:-1, n+1], -1, 1)

# Visualization: Velocity evolution at different times
time_intervals = np.arange(0, tmax + 0.25, 0.25)
time_indices = [min(int(ti / dt), nt - 1) for ti in time_intervals]  # Ensure indices are within range

fig, ax = plt.subplots(figsize=(8, 6))

# Plot u vs. t for specific positions at intervals of 0.25
for idx, t_idx in enumerate(time_indices):
    ax.plot(x, u[:, t_idx], label=f't={time_intervals[idx]:.2f}')

ax.set_xlabel('Position')
ax.set_ylabel('Velocity')
ax.legend()
ax.set_title('Velocity Profile at Different Time Intervals')

fig.tight_layout()
plt.show()
```
